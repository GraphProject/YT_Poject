```python
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

### 24 - Entertainment

```python
df = pd.read_csv("YT/CA_youtube_trending_data.csv")
keyword_list = []
for i in range(0,1000):
    if(df.loc[i, 'categoryId']==19):
        print('Loop No:'+ str(i))
        string = str(df.loc[i, 'tags']) + str(df.loc[i, 'description']) + str(df.loc[i, 'title'])
        print(string)
        keywords_all = keywords(string, words=30, lemmatize=True).split('\n')
        """
        for keyword in keywords_all:
            print('Video_ID :'+str(df.loc[i, 'video_id'])+'::Keyword::'+str(keyword)+'\n')

        """
        keyword_list.append(keywords_all)
```

```python
#RUN
keyword_list
```

```python
#RUN
count = 0
for keyword in keyword_list:
    print(count,end='')
    print(keyword)
    count = count + 1
```

```python
#RUN
keyword_marge = []
for keyword in keyword_list:
    keyword_marge = keyword_marge + keyword
keyword_marge
```

```python
comment_words = ''
STOPWORDS = ['https\'','https','http','http\'','video']+ list(STOPWORDS)
stopwords = set(STOPWORDS)
print(STOPWORDS)
#print(keyword_marge)
print(type(keyword_marge))
keyword_marge_string = str(keyword_marge)
print(type(keyword_marge_string))
   
word_cloud = WordCloud(
        width=3000,
        height=2000,
        random_state=1,
        background_color="salmon",
        colormap="Pastel1",
        collocations=False,
        stopwords=STOPWORDS,
        ).generate(keyword_marge_string)

# Display the generated Word Cloud
plt.figure(figsize = (20, 10), facecolor = None)
plt.imshow(word_cloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
